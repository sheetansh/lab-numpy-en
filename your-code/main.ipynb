```python
### [your code here]

print(a.shape)
print(b.shape)
print(""" As broadcasting is based on the spcific rule : 
      Either the Dimension of the 2 Matrices is Compatible ie same or 1. As here the Dimensions are not same. """)
print("But here are 3 ways to do it properly.")

# Expanding the given matrices : expanded_a and expanded_b

a_exp = np.expand_dims(a, axis=0)
b_exp = np.expand_dims(b, axis=-1)

a_broad,b_broad = np.broadcast_arrays(a_exp,b_exp)

result_expanded = np.add(a_broad,b_broad)
print(f'''\n So now the addition works as i am able to expand the Dimensions and add
      using the rules of Broadcasting attribute of numpy: \n {result_expanded}''')


# Adding new axis : 

a_newaxis = a[np.newaxis,:,:,:]
b_newaxis = b[:,:,:,np.newaxis]
result_newaxis = a_newaxis+b_newaxis

print(f'''\n So now the addition works as i am able to add a new axis and simply add
      without using the broadcasting attribute of numpy:\n {result_expanded}''')

# Adding by injecting a 'NONE' ie size 1 axis

a_size1axis = a[None,...]
b_size1axis = b[...,None]

result_size1axis = a_size1axis+b_size1axis

print(f'''\n So now the addition works as i am able to inject a size 1 axis and simply add
      using the broadcasting rules of numpy:\n {result_expanded}''')
```


### 12. Multiply a and c. Assign the result to e.


```python
### [your code here]
print("as the dimensions of the 2 matrices are not same ,we will use einsum ")
print(''' 
The Logic of np.einsum :
When you call np.einsum('ijk,mik->ijm', A, C), you are passing two things: 
a string of dimension labels ('ijk,mik->ijm') and the matrices (A, C).
The engine follows three strict rules to figure out what mathematical operation to perform:
1. Alignment (Multiplication): If a letter appears in both input strings (like i and k), einsum knows it must 
multiply the elements along those axes together. It iterates through the shared dimensions exactly like a nested for loop.
2. Summation (Reduction): If a letter appears in the inputs 
but is missing from the output string after the -> arrow (like k in our example), 
einsum knows it must sum the results of the multiplication along that axis. This is the mathematical definition of a dot product.
3. Preservation (Element-wise): If a letter appears in the inputs and also in the output string 
(like i, j, and m), einsum keeps that dimension in the final tensor without summing it away.
​ ''')
result_einsum_mult = np.einsum('ijk,mik->ijm',a,c)
e = result_einsum_mult
print(f"the final answer is : {e}")
```

If a value in d is larger than d_mean but smaller than d_max, assign 75 to the corresponding value in f.
If a value equals to d_mean, assign 50 to the corresponding value in f.
Assign 0 to the corresponding value(s) in f for d_min in d.
Assign 100 to the corresponding value(s) in f for d_max in d.
In the end, f should have only the following values: 0, 25, 50, 75, and 100.
Note: you don't have to use Numpy in this question.



```python
### [your code here]
conditions = [
    np.isclose(d,d_min),
    np.isclose(d,d_max),
    np.isclose(d,d_mean),
    (d>d_min) & (d<d_mean),
    (d>d_mean) & (d<d_max)
]

choices = [0,100,50,25,75]
f = np.select(conditions,choices,default=-1)
print(f"\nValue of f =\n {f}")

### Another way to do it : 
for x in range(0,len(d.flat)):

   if (d_mean < d.flat[x]) and (d.flat[x] < d_max):
       f.flat[x] = 75
   if d_min < d.flat[x] < d_mean:
       f.flat[x] = 25
   if d.flat[x] == d_mean:
       f.flat[x] = 50
   if d.flat[x] == d_min:
       f.flat[x] = 0
   if d.flat[x] == d_max:
       f.flat[x] = 100
    
print(f)
```
